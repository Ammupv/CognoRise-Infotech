```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV data into a DataFrame
cereal_df = pd.read_csv('/content/cereal.csv')
```

```python
# Select only the numeric columns from the DataFrame
numeric_columns = cereal_df.select_dtypes(include=np.number).columns
cereal_df_new = cereal_df[numeric_columns]
```

```python
# Create a new DataFrame containing only the numeric columns
cereal_corr = cereal_df_new.corr()
# Print the correlation matrix
print(cereal_corr)
```

```python
# ones_like can build a matrix of booleans (True, False) with the same shape as our data
ones_corr = np.ones_like(cereal_corr, dtype=bool)
ones_corr
```

```python
# checks the shapes of potentially unseen correlation matrices for debugging purposes.
ones_corr.shape, cereal_corr.shape
```

```python
# creates a mask using NumPy's triu function to exclude the lower triangle of the correlation matrix.
mask = np.triu(ones_corr)
mask
```

```python
# reate a heatmap using Seaborn, but it would include the redundant lower triangle.
sns.heatmap(data=cereal_corr, mask=mask);
```

```python
# creates a new mask that excludes both the upper and lower triangles of the correlation matrix,
# leaving only the non-redundant parts for the heatmap.
adjusted_mask = mask[1:, :-1]
adjusted_mask = mask[1:, :-1]
adjusted_mask
```

```python
adjusted_cereal_corr = cereal_corr.iloc[1:, :-1]
```

```python
# creates a new figure and axis for the heatmap, sets the figure size, and defines a colormap suitable
fig, ax = plt.subplots(figsize=(10,8))
cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)
# creates the heatmap using Seaborn. Here's a breakdown of the arguments:
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask,
            annot=True, annot_kws={"fontsize":13}, fmt=".2f", cmap=cmap,
            vmin=-1, vmax=1,
            linecolor='white', linewidths=0.5);

yticks = [i.upper() for i in adjusted_cereal_corr.index]
xticks = [i.upper() for i in adjusted_cereal_corr.columns]

ax.set_yticklabels(yticks, rotation=0, fontsize=13);
ax.set_xticklabels(xticks, rotation=90, fontsize=13);
title = 'CORRELATION MATRIX\nSAMPLED CEREALS COMPOSITION\n'
ax.set_title(title, loc='left', fontsize=18);
```
